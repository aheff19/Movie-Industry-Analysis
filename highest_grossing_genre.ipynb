```python
import pandas as pd
```

```python
# Assigning clean movie budgets to a variable
tn_movie_budgets_clean = pd.read_csv("tn_movie_budgets_clean.csv")
tn_movie_budgets_clean.head()
```

```python
#Assigning clean title basics to a variable
title_basics_cleaned = pd.read_csv("title_basics_cleaned.csv")
title_basics_cleaned.head()
```

```python
#Checking the shape of cleaned title basics and cleaned movie budgets
print(tn_movie_budgets_clean.shape)
print(title_basics_cleaned.shape)
```

```python
#Checking for unique title values
print(len(tn_movie_budgets_clean["movie"].unique()))
print(len(title_basics_cleaned["primary_title"].unique()))
print(len(title_basics_cleaned["original_title"].unique()))
```

```python
#Creating a dataframe where the primary and original title are in different languages
titles = title_basics_cleaned[title_basics_cleaned.primary_title != title_basics_cleaned.original_title]
titles.head()
titles.tail()
```

```python
#Creating a dataframe where the primary and original titles are the same
titles2 = title_basics_cleaned[title_basics_cleaned.primary_title == title_basics_cleaned.original_title]
titles2.tail()
```

```python
#Checking the columns of title_basics_cleaned
title_basics_cleaned.columns
```

```python
#Checking the columns of movie budgets cleaned
tn_movie_budgets_clean.columns
```

```python
#Creating a new dataframe from title_basics_cleaned containing only primary title and genre
genres_df = title_basics_cleaned[["primary_title", "genres"]]
genres_df.head()
```

```python
#Renaming the column primary_title to movie
genres_df = genres_df.rename(columns={'primary_title': 'movie'})
genres_df.head()
```

```python
#Setting movie column as index
genres_df_index = genres_df.set_index("movie")
genres_df_index.head()
```

```python
#Creating a new dataframe from tn_movie_budgets_clean containing only movie and worldwide_gross
budget_df = tn_movie_budgets_clean[["movie", "worldwide_gross"]]
budget_df.head()
```

```python
#Setting movie column as index
budget_df_index = budget_df.set_index("movie")
budget_df_index.head()
```

```python
#Merging budget df_index and genre_df_index inner
genres_budget_df = budget_df_index.join(genres_df_index, how='inner')
genres_budget_df.head()
genres_budget_df.tail()

```

```python
#reseting index
new_genres_budget_df = genres_budget_df.reset_index()
new_genres_budget_df.head()
```

```python
#Using value counts to check for movie duplicates
new_genres_budget_df["movie"].value_counts()
```

```python
#Creating a dataframe with rows corresponding only to "Home" value of movie column
home_df = new_genres_budget_df[new_genres_budget_df["movie"] == "Home"]
home_df.head()
```

```python
home_df["genres"].value_counts()
```

```python
#Creating a dataframe with rows corresponding only to "The Gift" value of movie column
gift_df = new_genres_budget_df[new_genres_budget_df["movie"] == "The Gift"]
gift_df.head()
```

```python
#Checking new_genres_budget_df for duplicates
duplicates1 = new_genres_budget_df[new_genres_budget_df.duplicated(subset='movie')]
print(len(duplicates1))
duplicates1.tail()
duplicates1.head()
```

```python
#Checking new_genres_budget_df for duplicates
duplicates2 = new_genres_budget_df[new_genres_budget_df.duplicated(subset='worldwide_gross')]
print(len(duplicates2))
duplicates2.tail()
duplicates2.head()
```

```python
#Dropping duplicates for column movie
result_df1 = new_genres_budget_df.drop_duplicates(subset='movie')
result_df1
```

```python
#Dropping duplicates for column worldwide_gross
result_df2 = result_df1.drop_duplicates(subset='worldwide_gross')
result_df2
```

```python
#Checking that there are no duplicates for movies using value counts
result_df2["movie"].value_counts()
```

```python
#Checking that there are no duplicates for worldwide_gross using value counts
result_df2["worldwide_gross"].value_counts()
```

```python
#Checking for missing values and drop first row
print(result_df2.isna().sum())
result_df2 = result_df2.drop(result_df2.index[0])
result_df2.head()
```

```python
#Identify genres
print(result_df2.shape)
print(len(result_df2["genres"].unique()))
```

```python
#Sort dataframe by worldwide_gross descending
result_df2 = result_df2.drop(["number_of_genres"], axis=1)
sorted_df = result_df2.sort_values(by='worldwide_gross', ascending=False)
sorted_df 
```

```python
#Sort dataframe by worldwide_gross ascending
sorted_df2 = result_df2.sort_values(by='worldwide_gross')
sorted_df2.head(15) 
```

```python
#Sort dataframe by genres ascending
sorted_df3 = result_df2.sort_values(by='genres')
sorted_df3.head(15) 
```

```python
#Group dataframe by genres
grouped_by_genres = sorted_df3.groupby('genres').mean()
grouped_by_genres
```

```python

type(grouped_by_genres["worldwide_gross"]["['Action', 'Adventure', 'Animation']"])
```

```python
#Sort grouped_by_genres in descending order
sorted_and_grouped = grouped_by_genres.sort_values(by='worldwide_gross', ascending=False)
sorted_and_grouped.head(15)
```

```python
#Converting column worldwide_gross from float to int
sorted_and_grouped["worldwide_gross"] = sorted_and_grouped["worldwide_gross"].apply(lambda x: int(x))
sorted_and_grouped.head()
```
